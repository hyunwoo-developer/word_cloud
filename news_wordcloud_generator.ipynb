```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests

html  = requests.get("https://www.yna.co.kr/news") # 원하는 언론사 선택
soup = BeautifulSoup(html.text, 'html.parser')
news = ""

title = soup.findAll('strong', {'class': 'tit-news'}) # 기사 타이틀 태그 지정

for i in title:
    news += i.text

wc = WordCloud(background_color='white', colormap='winter', font_path='AppleGothic').generate(news)
plt.figure(figsize=(20,15))
plt.imshow(wc)
plt.axis("off")
plt.show()
```
